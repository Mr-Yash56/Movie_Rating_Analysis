```python
## Objective
# Analyze rating distribution, genres, languages,
# year trends and correlations.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
# Load Clean Data
df = pd.read_csv("../data/clean_movies.csv")
```

```python
# Rating Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Rating(10)'], bins=20)
plt.title("Rating Distribution")
plt.show()
```

```python
# Top Movies
top_movies = df[df['Votes'] > 1000]
top_movies.sort_values('Rating(10)', ascending=False)[['Movie Name','Rating(10)','Votes']].head(10)
```

```python
# Genre Analysis
df['Genre'] = df['Genre'].str.split(',')
genre_df = df.explode('Genre')
genre_df['Genre'] = genre_df['Genre'].str.strip()
```

```python
# Average Rating
genre_rating = genre_df.groupby('Genre')['Rating(10)'].mean().sort_values(ascending=False)
genre_rating.head(10).plot(kind='bar')
plt.title("Top Genres by Average Rating")
plt.show()
```

```python
# Language Analysis
df.groupby('Language')['Rating(10)'].mean().sort_values(ascending=False)
df['Language'].value_counts()
```

```python
# Yearly Trends
year_rating = df.groupby('Year')['Rating(10)'].mean()

plt.figure(figsize=(10,5))
year_rating.plot()
plt.title("Average Rating Over Years")
plt.show()
```

```python
# Correlation
sns.heatmap(df[['Rating(10)','Votes','Timing(min)','Year']].corr(), annot=True)
plt.show()
```
